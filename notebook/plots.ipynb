```python
import matplotlib.pyplot as plt
```

```python
plt.bar(x = ['1n1c', '1n8c', '2n8c'], height=[7*60+30, 1*60+39, 1*60+52], color='teal');
plt.title('Execution time on different partitions')
plt.xlabel('Partition Type (n = nodes, c = cores)')
plt.ylabel('Execution Time (s)')
for i, v in enumerate([7*60+30, 1*60+39, 1*60+52]):
    plt.text(i-0.11, 
              v/[7*60+30, 1*60+39, 1*60+52][i]+30, 
              [7*60+30, 1*60+39, 1*60+52][i], 
              fontsize=15, 
              color='white')

plt.savefig('../plots/Execution_Time_Bar_Chart.jpg')
```
